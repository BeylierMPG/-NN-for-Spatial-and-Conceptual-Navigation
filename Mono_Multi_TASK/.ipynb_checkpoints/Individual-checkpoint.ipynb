```python
class Net_Individual(torch.nn.Module):
    def __init__(self,input_dimension,nodes_second,nodes_third):
        super(Net_Individual, self).__init__()

        self.fc1 = nn.Linear(in_features = input_dimension, out_features = nodes_second)
        self.fc2 = nn.Linear(in_features = nodes_second, out_features = nodes_third)
        self.fc3 = nn.Linear(in_features = nodes_third,  out_features = 1)
        
        #self.fc1 = nn.Linear(4, 64) 
        #self.fc2 = nn.Linear(64, 64)
        #self.fc3 = nn.Linear(64, 1)
        
        self.relu = nn.ReLU()
        
        
        #torch.nn.init.xavier_normal_(self.fc1.weight)
        #torch.nn.init.xavier_normal_(self.fc2.weight)
        #torch.nn.init.xavier_normal_(self.fc3.weight)
        
       

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.fc2(x) #tanh if does not work than not a smoothness issue
        x = self.fc3(x)
        return x


    
    
class Training_individual():

    # The class initialisation function.
    def __init__(self,model,opti,loss):
        self.model = model
        self.optimizer = opti
        self.criterion = loss

    
    def train_network(self,data,target,pumpkin_seed,multi_task):


        Loss = []  
        #Compute the prediction
        output = self.model.forward(data)

        if len(output.shape) >len(target.shape):
            output = output.squeeze(len(output.shape)-1)
        # Calculate the loss for this transition.

        output_loss = self.criterion(output,target)
        #print("output",output)
        #print("target",target)
       # print("loss",loss)
        # Set all the gradients stored in the optimiser to zero.
        self.optimizer.zero_grad()

        output_loss.backward() 

        # Take one gradient step to update the Q-network. 
        self.optimizer.step()

        return output_loss
    
    
    
    

```

```python
SPLITS = 2

LOSSES_LR = [[]for i in range(SPLITS)]
LOSSES_LR_val = [[]for i in range(SPLITS)]
Fold = 0
L_model = [[]for i in range(SPLITS)]

kf = KFold(n_splits=SPLITS, random_state=None, shuffle=False)
        
for train_index, test_index in kf.split(dataset_space):
    print(f'Fold number {Fold}')
    print('--------------------------------')
    print("TRAIN:", len(train_index), "TEST:", len(test_index))
    
    train_subsampler = torch.utils.data.SubsetRandomSampler(train_index)
    test_subsampler = torch.utils.data.SubsetRandomSampler(test_index)
    
    train_loader = torch.utils.data.DataLoader(dataset_space,batch_size=16,shuffle = False ,sampler=train_subsampler)
    test_loader = torch.utils.data.DataLoader(dataset_space,batch_size=16,shuffle = False , sampler=test_subsampler)
       
    model = Net_Individual(input_dimension=4,nodes_second = 6,nodes_third = 4)    
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001, betas=(0.9, 0.999), eps=1e-08, weight_decay=0, amsgrad=False)
    criterion = nn.BCEWithLogitsLoss()


    losses = []
    val_losses = []

    for epoch in range(EPOCH):

        loss = 0
        val_loss = 0 

        for x_batch, y_batch in train_loader:

            x_batch = x_batch[:,0:4]
            
            optimizer.zero_grad()
            output = model(x_batch)

            output_loss = criterion(output.squeeze(1),y_batch)
            output_loss.backward()
            optimizer.step()
            
            loss += output_loss.detach().numpy()
        losses.append(loss/len(train_loader))
            
        with torch.no_grad():
            for x_val, y_val in test_loader:
                model.eval() # not useful for now
                x_val = x_val[:,0:4]
                yhat = model(x_val).squeeze(1)
                val_loss += criterion(yhat,y_val)
            val_losses.append(val_loss.item()/len(test_loader))


        if np.mod(epoch,20)==0: 
            print("Epoch {}, val_loss {}".format(epoch, val_loss.item()/len(test_loader))) 
            fig = plt.figure()
            ax = fig.add_subplot(111)
            plt.plot(np.arange(len(val_losses)), val_losses)
            plt.ylabel('Loss')
            plt.xlabel('Epoch #')
            plt.show()
            

    LOSSES_LR[Fold] = losses
    
    LOSSES_LR_val[Fold] = val_losses
    L_model[Fold] = model
    Fold+=1
    print(val_losses[-1])

l_epoch = [i for i in range(len(losses))]
l_epoch_val = [i for i in range(len(val_losses))]
liste_label = [np.str(i) for i in range(SPLITS)]

plt.figure(figsize=(20, 10))
for i in range(Fold):
    plt.plot(l_epoch_val,LOSSES_LR_val[i], '--', linewidth=1,label = liste_label[i])
    
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(loc='center', bbox_to_anchor=(0.5, -0.30), shadow=False, ncol=Fold)
plt.title("WEATHER Task")
```
